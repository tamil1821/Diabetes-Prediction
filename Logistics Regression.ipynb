```python
con_data=data[['Pregnancies','Outcome']]
con_data
```

```python
type(data1)
```

```python
data2=pd.DataFrame(data1,columns=['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'])
data2
```

```python
final_df=pd.concat([data2,con_data],axis=1)
final_df
```

```python
# FEATRURE SELECTION


sns.heatmap(data2.corr(),annot=True)
plt.show()
```

```python
MODEL CRTEATION
```

```python
# step 1 Crearting independent and dependent varaiable

X=final_df.iloc[:,:-1]
y=final_df.Outcome
y
```

```python
# step 2 Creating training and testing data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=45)
y_test
```

```python
# step 3 Model craetion
# object creation
# fit---Traning
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(X_train,y_train)
```

```python
# step 4 Predection
y_pred=clf.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
recall=recall_score(y_test,y_pred)
recall
```

```python
precision=precision_score(y_test,y_pred)
precision
```

```python
accuracy=-accuracy_score(y_test,y_pred)
accuracy
```

```python
f1score=f1_score(y_test,y_pred)
f1score
```

```python
cr=classification_report(y_test,y_pred)
print(cr)
```
